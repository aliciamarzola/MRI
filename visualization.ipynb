```python
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
```

```python
brain_t1c = nib.load('datasets/brats/training_data/BraTS-GLI-02405-100/BraTS-GLI-02405-100-t1c.nii.gz')
brain_t1c_data = brain_t1c.get_fdata()
```

```python
affine = brain_t1c.affine
shape = brain_t1c.shape
```

```python
print(affine)
```

```python
print(shape)
```

```python
brain_t1c.header.get_zooms()
```

```python
nib.aff2axcodes(affine)
```

```python
fig, axis = plt.subplots(1,3)
axis[0].imshow(brain_t1c_data[shape[2]//2,:,:], cmap='gray')
axis[1].imshow(brain_t1c_data[:,:,shape[2]//2], cmap='gray')
axis[2].imshow(brain_t1c_data[:,shape[2]//2,:], cmap='gray')

for ax in axis:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
